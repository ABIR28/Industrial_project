```python
# to preces dataframe function
def processing(df):
    df.rename(columns=df.iloc[0], inplace = True)
    df.drop(df.index[0], inplace = True)
    df.drop(df.columns[1],axis = 1 ,inplace = True)
    df.reset_index(drop=True, inplace=True)
    #data.rename(columns = {'LOX Net Production (ton)': 'LOX_Net', 'GAN Net Production (ton)': 'GAN_Net', 'LIN Net Production(ton)':'LIN_Net', 'Total Net liquid production (ton)':'Total_Net_liquid'}, inplace = True)
    df['Date'] = pd.to_datetime(df['Date']).dt.normalize()
    return df
```

```python
df2 = df.copy()
df2.sample(5)
```

```python
dataframe = pd.DataFrame(columns=['Date','production_1','production_2','production_3','Total production (ton)'])
st=0
for i in range(12):
    df1 = df2.iloc[st:st+6,:].T
    df1 = processing(df1)
    dataframe= dataframe.append(df1,ignore_index=True)
    st= st+11

    
dataframe.head()  
```

```python
#detect ouliers drop 
from collections import Counter

def detect_outliers(dataframe, n, features):
    outlier_indices = [] 
    for col in features: 
        Q1 = np.percentile(dataframe[col], 25)
        Q3 = np.percentile(dataframe[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR 
        outlier_list_col = dataframe[(dataframe[col] < Q1 - outlier_step) | (dataframe[col] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col) 
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(key for key, value in outlier_indices.items() if value > n) 
    return multiple_outliers

outliers_to_drop = detect_outliers(dataframe, 2, ['production_1', 'production_2', 'production_3', 'Total production (ton)'])
print("We will drop these {} indices: ".format(len(outliers_to_drop)), outliers_to_drop)
```

```python
# Outliers in numerical variables

dataframe.loc[outliers_to_drop, :]
```

```python
# Drop outliers and reset index

print("Before: {} rows".format(len(dataframe)))
dataframe = dataframe.drop(outliers_to_drop, axis = 0).reset_index(drop = True)
print("After: {} rows".format(len(dataframe)))
```
